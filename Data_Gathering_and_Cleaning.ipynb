**Inplace Attribute**

```python
import pandas as pd

data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35]}
df = pd.DataFrame(data)
df_dropped = df.drop(columns='Age')   #default value is False for Inplace
print(df)  # The original DataFrame remains unchanged
print(df_dropped)  # A new DataFrame without the 'Age' column
print(df)#The original DataFrame remains unchanged
```

```python
import pandas as pd

data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35]}
df = pd.DataFrame(data)
df_dropped = df.drop(columns='Age')   #default value is False for Inplace
print(df)  # The original DataFrame remains unchanged
print(df_dropped)  # A new DataFrame without the 'Age' column

# Using inplace=True
df.drop(columns='Age', inplace=True)
print(df)
```
